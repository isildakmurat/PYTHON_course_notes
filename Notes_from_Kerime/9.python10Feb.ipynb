```python
şehirler = ["istanbul", "van", "ağrı", "ankara"]
```

```python
sorted(şehirler)  # alrabetik sıraya göre sırala dedik
```

```python
sorted(şehirler, key = len)  # sıralama emrine "key" ile yeni parametre atayıp 
                            #onu len (uzunluk) yaptık, böylelikle uzunluğa göre sıralayacak
```

```python
sorted(şehirler, key = len, reverse = True)  # bir öncekini tersine sıralama emri verdik
```

```python
şehirler.sort(key = len)
şehirler
```
